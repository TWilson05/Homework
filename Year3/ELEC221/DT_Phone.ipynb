```python
def dial_phone(phone_number, sample_length=1, sample_rate=44100):
    """Creates a sequence of sound waveforms that are equivalent
    to dialing the provided phone number.

    Args:
        phone_number (str): A string representing a phone number. You
            can assume this is guaranteed to be in the format "XXX-XXXX".
        sample_length (float): The length, in seconds, of the dial tone.
        sample_rate (int): The number of samples per second to take for
            the dial tone.

    Returns:
        list[array]: A list of numpy arrays, one per digit, that consist of
        the dialtone provided by each number. Each array should have length
        sample_length * sample_rate.
    """
    tone_range = np.arange(0,sample_length*sample_rate)
    output = []
    for digit in phone_number:
      if(digit == "1"):
        tone = np.sin(2 * np.pi * 697 * tone_range / sample_rate) + np.sin(2 * np.pi * 1209 * tone_range / sample_rate)
        output.append(tone)
      if(digit == "2"):
        tone = np.sin(2 * np.pi * 697 * tone_range / sample_rate) + np.sin(2 * np.pi * 1336 * tone_range / sample_rate)
        output.append(tone)
      if(digit == "3"):
        tone = np.sin(2 * np.pi * 697 * tone_range / sample_rate) + np.sin(2 * np.pi * 1477 * tone_range / sample_rate)
        output.append(tone)
      if(digit == "4"):
        tone = np.sin(2 * np.pi * 770 * tone_range / sample_rate) + np.sin(2 * np.pi * 1209 * tone_range / sample_rate)
        output.append(tone)
      if(digit == "5"):
        tone = np.sin(2 * np.pi * 770 * tone_range / sample_rate) + np.sin(2 * np.pi * 1336 * tone_range / sample_rate)
        output.append(tone)
      if(digit == "6"):
        tone = np.sin(2 * np.pi * 770 * tone_range / sample_rate) + np.sin(2 * np.pi * 1477 * tone_range / sample_rate)
        output.append(tone)
      if(digit == "7"):
        tone = np.sin(2 * np.pi * 852 * tone_range / sample_rate) + np.sin(2 * np.pi * 1209 * tone_range / sample_rate)
        output.append(tone)
      if(digit == "8"):
        tone = np.sin(2 * np.pi * 852 * tone_range / sample_rate) + np.sin(2 * np.pi * 1336 * tone_range / sample_rate)
        output.append(tone)
      if(digit == "9"):
        tone = np.sin(2 * np.pi * 852 * tone_range / sample_rate) + np.sin(2 * np.pi * 1477 * tone_range / sample_rate)
        output.append(tone)
      if(digit == "0"):
        tone = np.sin(2 * np.pi * 941 * tone_range / sample_rate) + np.sin(2 * np.pi * 1336 * tone_range / sample_rate)
        output.append(tone)
    return output

# sample_length = 1
# sample_rate = 44100
# tone_range = np.arange(0, sample_length * sample_rate)
# tone = np.sin(2 * np.pi * 697 * tone_range / sample_rate) + np.sin(2 * np.pi * 1209 * tone_range / sample_rate) # 440 Hz
# Audio(tone, rate=sample_rate)

sample_rate = 44100
tone = dial_phone("321-4567")
print(len(tone))
Audio(tone[0],rate=sample_rate)
```

```python
def parse_phone_number(dial_tones, sample_length=1, sample_rate=44100):
    """Use Fourier analysis to determine the frequencies that are present
    in each tone, and return the phone number.

    Args:
        dial_tones (list[array]): A list of seven NumPy arrays, each representing
            one part of the phone number.
        sample_length (float): The length, in seconds, of the dial tone.
        sample_rate (int): The number of samples per second to take for
            the dial tone.

    Returns:
        str: The phone number indicated by dial_tones. This should be returned
        as a string with the form "XXX-XXXX"
    """
    def max_freq(tone):
      cks = np.fft.fft(tone)
      cks = np.absolute(cks)
      half = int(len(cks)/2)
      cks = cks[0:half]
      maxFreq = []
      maxFreq.append(np.argmax(cks))
      cks[maxFreq] = 0
      maxFreq.append(np.argmax(cks))
      return maxFreq

    def find_number(freq):
      num = "0"
      if freq[0] == 697 or freq[1] == 697:
        if freq[0] == 1209 or freq[1] == 1209:
          num = "1"
        if freq[0] == 1336 or freq[1] == 1336:
          num = "2"
        if freq[0] == 1477 or freq[1] == 1477:
          num = "3"
      if freq[0] == 770 or freq[1] == 770:
        if freq[0] == 1209 or freq[1] == 1209:
          num = "4"
        if freq[0] == 1336 or freq[1] == 1336:
          num = "5"
        if freq[0] == 1477 or freq[1] == 1477:
          num = "6"
      if freq[0] == 852 or freq[1] == 852:
        if freq[0] == 1209 or freq[1] == 1209:
          num = "7"
        if freq[0] == 1336 or freq[1] == 1336:
          num = "8"
        if freq[0] == 1477 or freq[1] == 1477:
          num = "9"
      if freq[0] == 941 or freq[1] == 941:
        num = "0"
      return num

    number = ""
    for i in range(0,7):
      number += find_number(max_freq(dial_tones[i]))
      if i == 2:
        number += "-"

    return number
tone = dial_phone("321-4567")
print(parse_phone_number(tone))
```
